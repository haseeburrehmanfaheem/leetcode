maximum subarray sum


2 ways to solve it
1- calculate the prefix sum, keep track of the minimum prefix sum seen so far, and then keep subtracting from the curr prefix sum
2- kadanes algorithm

```python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:

        prefix_sum = [0] * len(nums)
        prefix_sum[0] = nums[0]
        curr_min = min(0,nums[0])
        curr_max = nums[0]
        for i in range(1,len(nums)):
            prefix_sum[i] = prefix_sum[i-1] + nums[i]
            curr_max = max(curr_max, prefix_sum[i]-curr_min)
            curr_min = min(curr_min,prefix_sum[i])
        return curr_max

        curr_min = 0
        curr_max = prefix_sum[0]
        for each in prefix_sum:
            curr_max = max(curr_max, each-curr_min)
            curr_min = min(curr_min, each)
        
        return curr_max



# easier way to solve it using prefix 
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        curr_min,curr_max = 0, nums[0]
        prefix = 0
        for each in nums:
            prefix += each
            curr_max = max(curr_max,prefix-curr_min)
            curr_min = min(curr_min,prefix)
        return curr_max

        
```

Jumpgame2, similar but minimum

## kindoff bfs

```python
class Solution:
    def jump(self, nums: List[int]) -> int:
        l ,r = 0,0
        jumps = 0
        while r < len(nums)-1:
            maxJump = 0
            for i in range(l,r+1):
                maxJump = max(maxJump,nums[i]+i)
            l=r+1
            r=maxJump
            jumps+=1
        return jumps
```
