```python
# ok so this guy is wierd to implement

# since the intervals are already sorted, we have to do it using one pass
# otherwise, we could have sorted and merged in place, like we did the above question]


# so in this case, if we can add the interval before an existing interval, best case scenario, we add and add rest and return
# otherwise, we keep updating our merge interval, and then merge at the end

class Solution:
    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:

        merged = []
        for i in range(len(intervals)):
            if newInterval[1] < intervals[i][0]:
                merged.append(newInterval)
                merged += intervals[i:]
                return merged
            
            elif newInterval[0] > intervals[i][1]:
                merged.append(intervals[i])
                continue
            else:
                newInterval[0] = min(newInterval[0],intervals[i][0])
                newInterval[1] = max(newInterval[1],intervals[i][1])
        
        merged.append(newInterval)
        return merged
   
```
