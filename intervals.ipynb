```python
# ok so this guy is wierd to implement

# since the intervals are already sorted, we have to do it using one pass
# otherwise, we could have sorted and merged in place, like we did the above question


# so in this case, if we can add the interval before an existing interval, best case scenario, we add and add rest and return
# otherwise, we keep updating our merge interval, and then merge at the end

class Solution:
    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:

        merged = []
        for i in range(len(intervals)):
            if newInterval[1] < intervals[i][0]:
                merged.append(newInterval)
                merged += intervals[i:]
                return merged
            
            elif newInterval[0] > intervals[i][1]:
                merged.append(intervals[i])
                continue
            else:
                newInterval[0] = min(newInterval[0],intervals[i][0])
                newInterval[1] = max(newInterval[1],intervals[i][1])
        
        merged.append(newInterval)
        return merged
   
```

```python
class Solution:
    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        intervals.sort()

        end = intervals[0][1]
        count = 0
        # we know that we should take the interval with greater end time, and remove it, since it will cost the most trouble
        for i in range(1,len(intervals)):
            if end > intervals[i][0]: # so try taking the minimum one and add 1
                end = min(intervals[i][1], end)  
                count += 1
            else: # otherwise, just update the mini
                end = intervals[i][1]
        return count
                
        
```

```python
"""
Definition of Interval:
class Interval(object):
    def __init__(self, start, end):
        self.start = start
        self.end = end
"""

class Solution:
    def canAttendMeetings(self, intervals: List[Interval]) -> bool:

        intervals.sort(key=lambda i: i.start)

        for i in range(1,len(intervals)):
            if intervals[i].start < intervals[i-1].end:
                return False
        return True


```

```python
# tough question
# trick is, how to keep track of which room is free
# and we can use a min heap, and have the end times as the value
# whatever room gets freed first, we can assign the meeting in that room


"""
Definition of Interval:
class Interval(object):
    def __init__(self, start, end):
        self.start = start
        self.end = end
"""

class Solution:
    def minMeetingRooms(self, intervals: List[Interval]) -> int:
        if not intervals:
            return 0
        intervals.sort(key=lambda i: i.start)
        heap = [intervals[0].end]
        total = len(heap)
        for i in range(1,len(intervals)):
            while heap and heap[0] <= intervals[i].start:
                heapq.heappop(heap)
            
            heapq.heappush(heap, intervals[i].end)
            total = max(total,len(heap))
        return total



        
```
