Simple BFS counting island

```python
# bfs solution

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:

        directions = [(1,0),(-1,0), (0,1), (0,-1)]
        def bfs(i,j):
            queue = deque()
            queue.append((i,j))
            while queue:
                i,j = queue.popleft()
                for di,dj in directions:
                    ni = i+di
                    nj = j+dj
                    valid_i = 0<=ni<len(grid)
                    valid_j = 0<=nj<len(grid[0])
                    if valid_i and valid_j and grid[ni][nj] == "1":
                        grid[ni][nj] = "0" # now visited
                        queue.append((ni,nj))
            
        islands = 0
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == "1":
                    islands += 1
                    bfs(i,j)
        return islands

    
    def numIslands(self, grid: List[List[str]]) -> int:
        def dfs(i,j):
            if i<0 or j<0 or i>=len(grid) or j>=len(grid[0]) or grid[i][j] == "0":
                return
            grid[i][j] = "0"
            dfs(i+1,j)
            dfs(i-1,j)
            dfs(i,j+1)
            dfs(i,j-1)
        
        islands = 0
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == "1":
                    islands += 1
                    dfs(i,j)
        return islands

        
```

Clone graph

```python
# dfs and bfs
class Solution:
    def cloneGraph(self, node: Optional['Node']) -> Optional['Node']: # dfs
        if node is None:
            return None
        map = {}
        def dfs(original):
            if original in map:
                return map[original] 
            curr = Node(original.val)
            map[original] = curr
            for each in original.neighbors:
                curr.neighbors.append(dfs(each))    
            return curr
        return dfs(node)
    
    def cloneGraph(self, node: Optional['Node']) -> Optional['Node']: # bfs
        visited = {}
        if not node:
            return node
        queue = deque()
        queue.append(node)
        visited[node] = Node(node.val)

        while queue:
            curr = queue.popleft()
            for neighbor in curr.neighbors:
                if neighbor not in visited:
                    visited[neighbor] = Node(neighbor.val)
                    queue.append(neighbor)
                visited[curr].neighbors.append(visited[neighbor])


        return visited[node]
```

Rotten Oranges : multi source bfs

```python
class Solution:
    def orangesRotting(self, grid: List[List[int]]) -> int:

        m,n = len(grid), len(grid[0])
        directions = [(1,0),(-1,0),(0,1),(0,-1)]
        queue = deque()


        def bfs(fresh):
            max_time = 0
            while queue:
                i,j,currtime = queue.popleft()

                for new_i,new_j in directions:
                    change_i,change_j = i+new_i,j+new_j
                    valid = (0<=change_i<m) and (0<=change_j<n)
                    if valid and grid[change_i][change_j] == 1:
                        max_time = currtime+1
                        grid[change_i][change_j] = 2
                        queue.append((change_i,change_j,max_time))
                        fresh -= 1
            return max_time if fresh == 0 else -1
        
        fresh = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 2:
                    queue.append((i,j,0))
                elif grid[i][j] == 1:
                    fresh+=1
        
        return 0 if fresh == 0 else bfs(fresh)
        
```

Pacific Atlantic

set intersection

```python
class Solution:
    def pacificAtlantic(self, heights: List[List[int]]) -> List[List[int]]:

        pacific_set = set()
        atlantic_set = set()
        m,n = len(heights),len(heights[0])
        directions = [(1,0),(-1,0),(0,1),(0,-1)]
        def bfs(queue,visited):
            while queue:
                i,j = queue.popleft()
                for di,dj in directions:
                    ni = i+di
                    nj = j+dj
                    valid = (0<=ni<m) and (0<=nj<n)
                    if valid and heights[ni][nj] >= heights[i][j] and (ni,nj) not in visited:
                        visited.add((ni,nj))
                        queue.append((ni,nj))
            return
        
        pacific_queue = deque()
        atlantic_queue = deque()
        pacific_set = set()
        atlantic_set = set()
        for i in range(m):
            pacific_queue.append((i,0))
            pacific_set.add((i,0))
            atlantic_queue.append((i,n-1))
            atlantic_set.add((i,n-1))


        for j in range(n):
            pacific_queue.append((0,j))
            pacific_set.add((0,j))
            atlantic_queue.append((m-1,j))
            atlantic_set.add((m-1,j))

        bfs(pacific_queue, pacific_set)
        bfs(atlantic_queue, atlantic_set)
        inter = atlantic_set.intersection(pacific_set) 

        return [[i,j] for i,j in inter]


        
```

## Topological Sort

```python
# adj list[prereq] = [courses that require prereq to be completed]
# indegree[course] -> prereq required to complete the course


class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        adj_list = defaultdict(list)
        indegree = [0]*numCourses
        for after,before in prerequisites:
            adj_list[before].append(after)
            indegree[after] += 1
        
        queue = deque()

        for i in range(numCourses):
            if indegree[i] == 0:
                queue.append(i)

        

        processed = 0
        while queue:
            course_taken = queue.popleft()
            processed += 1
            for nei in adj_list[course_taken]:
                indegree[nei]-= 1
                if indegree[nei] == 0:
                    queue.append(nei)
        
        if processed == numCourses:
            return True

        return False
```

```python
class Solution:
    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        adj_list = defaultdict(list)
        indegree = [0] * numCourses
        for after,before in prerequisites:
            adj_list[before].append(after)
            indegree[after] += 1

        queue = deque()
        for i in range(numCourses):
            if indegree[i]==0:
                queue.append(i)


        processed = 0
        res = []
        while queue:
            curr = queue.popleft()
            processed += 1
            res.append(curr)
            for nei in adj_list[curr]:
                indegree[nei]-=1
                if indegree[nei]==0:
                    queue.append(nei)

        return res if processed == numCourses else []
```

Valid Graph

do dfs and count visited or cycles
or do union-find -> edges should be n-1, and each edge should connect components, otherwise we are introducing a cycle

```python
class Solution(object):
    def validTree(self, n, edges):
        """
        :type n: int
        :type edges: List[List[int]]
        :rtype: bool
        """
        if len(edges) != n - 1:
            return False

        adj = defaultdict(list)
        for f, s in edges:
            adj[f].append(s)
            adj[s].append(f)
        
        visited = set()
        def dfs(node,prev):
            if node in visited:
                return False
            visited.add(node)
            for nei in adj[node]:
                if nei == prev:
                    continue
                if not dfs(nei,node):
                    return False
            return True

        return dfs(0,None) and len(visited) == n
```

Number of connected components in graph

```python
# 3 ways, dfs, bfs, union find

# for dfs, try doing bfs on nodes, increase if a node is unvisited. Same for bfs
# lets try to write a union find algorithm

class Solution:
    def countComponents(self, n: int, edges: List[List[int]]) -> int:
        rank = [1] * n
        parent = [i for i in range(n)]
        def find(e1):
            while parent[e1] != e1:
                parent[e1] = parent[parent[e1]]
                e1 = parent[e1]
            return e1

        def union(e1,e2):
            p1,p2 = find(e1),find(e2)

            if p1 == p2:
                return False
            else:
                if rank[p1] >= rank[p2]:
                    rank[p1] += rank[p2]
                    parent[p2] = p1
                else:
                    rank[p2] += rank[p1]
                    parent[p1] = p2
                return True

        connected = n 
        for src,dst in edges:
            if union(src,dst):
                connected -= 1
        
        return connected
```

Find redundant connections
## unionfind

```python
class Solution:
    def findRedundantConnection(self, edges: List[List[int]]) -> List[int]:

        rank = [1] * (len(edges)+1)
        parent = [i for i in range((len(edges)+1))]

        def find(e1):
            while e1 != parent[e1]:
                parent[e1] = parent[parent[e1]]
                e1 = parent[e1]
            return e1
        
        def union(e1,e2):
            p1,p2 = find(e1),find(e2)

            if p1 == p2:
                return False
            else:
                if rank[p1] > rank[p2]:
                    rank[p1] += rank[p2]
                    parent[p2] = p1
                else:
                    rank[p2] += rank[p1]
                    parent[p1] = p2
                return True
        
        for e1,e2 in edges:
            if not union(e1,e2):
                return [e1,e2]
        
```

## Level by level bfs

To solve dijkstra with negative weights, we use belman ford algorithm
For the cheapest path within k stops
we need to take care of both, stops and cheapest path, therefor we can use level by level bfs traversal

```python
class Solution:
    def findCheapestPrice(self, n: int, flights: List[List[int]], src: int, dst: int, k: int) -> int:
        
        adjList = defaultdict(list)
        for u,v,w in flights:
            adjList[u].append((v,w))
        visited = [float("inf")] * n
        visited[src] = 0
        queue = deque()
        queue.append((src,0))
        hops = 0
        while queue and hops < k+1:
            size = len(queue)
            for i in range(size):
                node,dist = queue.popleft()
                for nei,cost in adjList[node]:
                    if cost+dist < visited[nei]:
                        visited[nei] = cost+dist
                        queue.append((nei,cost+dist))
            hops+=1
        return visited[dst] if visited[dst] != float("inf") else -1
        
```

## Reconstruct Itenary 

You are given a list of airline tickets where tickets[i] = [fromi, toi] represent the departure and the arrival airports of one flight. Reconstruct the itinerary in order and return it.

All of the tickets belong to a man who departs from "JFK", thus, the itinerary must begin with "JFK". If there are multiple valid itineraries, you should return the itinerary that has the smallest lexical order when read as a single string.

For example, the itinerary ["JFK", "LGA"] has a smaller lexical order than ["JFK", "LGB"].

# Find Eulerian Path
## actually just dfs with backtracking


```python
class Solution:
    def findItinerary(self, tickets: List[List[str]]) -> List[str]:
        tickets.sort()
        adjList = defaultdict(list)
        for src,dst in tickets:
            adjList[src].append(dst)
        
        res = ["JFK"]
        def dfs(src):
            if len(res) == len(tickets)+1:
                return True
            
            if len(adjList[src]) == 0:
                return False
            
            tmp = adjList[src].copy()
            for index,dst in enumerate(tmp):
                if index > 0 and tmp[index-1] == tmp[index]:
                    continue
                res.append(dst)
                adjList[src].pop(index)
                if dfs(dst):
                    return True
                res.pop()
                adjList[src].insert(index,dst)
            return False
        dfs("JFK")
        return res
```

## Minimum Spanning Tree
## Prims Algorithm

```python
class Solution:
    def minCostConnectPoints(self, points: List[List[int]]) -> int:
        n = len(points)
        visited = [False] * n
        heap = [(0,0)] # dist, node
        connected = 0
        cost = 0
        while connected < n:
            dist,node = heapq.heappop(heap)
            if visited[node]:
                continue
            visited[node] = True
            connected += 1
            cost += dist
            for i in range(n):
                if visited[i]:
                    continue
                tmpCost = abs(points[node][0] - points[i][0]) + abs(points[node][1] - points[i][1])
                heapq.heappush(heap,(tmpCost,i))
        
        return cost
        
        
        
```
