```python
# keep spiraling, go square by square


class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """


        l = 0
        r = len(matrix)-1

        while l<r:

            for i in range(r-l):
                # top left to top right
                tr = matrix[l+i][r]
                matrix[l+i][r] = matrix[l][l+i]  

                #top right to bottom right
                br = matrix[r][r-i]
                matrix[r][r-i] = tr

                # bottom right to bottom left
                bl = matrix[r-i][l]
                matrix[r-i][l] = br

                # bottom left to top left
                matrix[l][l+i] = bl
            l+=1
            r-=1
        
        return matrix
```

```python
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """

        m,n = len(matrix), len(matrix[0])
        row_set = set()
        col_set = set()


        for i in range(m):
            for j in range(n):

                if matrix[i][j] == 0:
                    row_set.add(i)
                    col_set.add(j)
        

        for i in row_set:
            for j in range(n):
                matrix[i][j] = 0
        

        for j in col_set:
            for i in range(m):
                matrix[i][j] = 0

        return 

        # O (3n) space O(m+n)
        # can improve space
        # we can store what row and col to mark 0 in the first row and col
        # need to handle the case for matrix[0][0] since we dont know if we have to mark the row 0 or col 0
        # intersection case
        #basically our set that we used above is going to be in our matrix now
    def setZeroes(self, matrix: List[List[int]]) -> None:
        rowzero = False
        m,n = len(matrix), len(matrix[0])
        for i in range(m):
            for j in range(n):

                if matrix[i][j] == 0:
                    
                    if i == 0:
                        rowzero = True
                    else:    
                        matrix[i][0] = 0
                    
                    matrix[0][j] = 0
        


        # setting rows
        for i in range(1,m):
                if matrix[i][0] == 0:
                    for j in range(n):
                        matrix[i][j] = 0
        
        # setting columns
        for j in range(0,n):
            if matrix[0][j] == 0:
                for i in range(m):
                    matrix[i][j] = 0

        if rowzero:
            for j in range(n):
                matrix[0][j] = 0


        return 
```
