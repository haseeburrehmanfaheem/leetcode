```python
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        
        # If we have next smallest element and previous smallest element
        # For each index
        # we can easily compute the "valid" area for each point


        n = len(heights)
        pse = [-1] * n
        stack = [(heights[0],0)] # val,index
        for i in range(1,n): 
            while stack and stack[-1][0] >= heights[i]:
                stack.pop()
            if stack:
                pse[i] = stack[-1][1]
            stack.append((heights[i],i))
        
        nse = [n] * n
        stack = [(heights[-1],n-1)]
        for i in range(n-2,-1,-1):
            while stack and stack[-1][0] >= heights[i]:
                stack.pop()
            if stack:
                nse[i] = stack[-1][1]
            stack.append((heights[i],i))
        

        area = 0
        for i in range(n):
            length = nse[i] - pse[i] - 1
            area = max(area, length * heights[i])
        

        return area
        




```

```python
# decode string is difficult
# revisit
# idea is 
# whenever we see [ , we will push whatever we have currently in our prev stack
# whenever we see ], we will multiply our curr string (after [) and add it to prev.pop(), and that will become our curr


class Solution:
    def decodeString(self, s: str) -> str:
        num_stack = []
        prev = []
        curr = []
        num = 0
        for each in s:
            if each.isdigit():
                # if digit, we keep multiplying
                num = num * 10
                num += int(each)
            elif each == "[":
                # whenever we see [, we push the num
                num_stack.append(num)
                prev.append(curr)
                curr = []
                num = 0
            elif each == "]":
                # if we see end, we pop from num stack, alpha stack and multiply
                multiply = num_stack.pop()
                final_str = prev.pop() + (curr * multiply)
                curr = final_str
            
            else:
                curr.append(each)
            # whenever we see ], we pop the num and the string from alph, and multiply with them

        return "".join(curr)
```
