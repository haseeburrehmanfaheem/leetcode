Climb Stairs 

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        res = [0] * (n+1)
        res[0] = 1
        res[1] = 1


        for i in range(2,n+1):
            res[i] = res[i-1] + res[i-2]
        
        return res[n]
        
```

Min Cost Climb Stairs

```python
def minCostClimbingStairs(self, cost: List[int]) -> int:
    min_cost = [0] * (len(cost))

    min_cost[0] = cost[0]
    min_cost[1] = cost[1] # since min(min_cost[0],min_cost[-1]) is 0

    for i in range(2, len(cost)):
        min_cost[i] = cost[i] + min(min_cost[i-1],min_cost[i-2])
    
    return min(min_cost[-1],min_cost[-2])
```

***Unique Paths simple mxn matrix***

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        # top down, memoization
        dp = [[-1 for _ in range(n)] for _ in range(m)] 
        def dfs(i,j):
            if i == 0 and j == 0:
                return 1
            
            elif i < 0 or j < 0:
                return 0
            
            if dp[i][j] != -1:
                return dp[i][j]
            up = dfs(i-1,j)
            left = dfs(i,j-1)
            dp[i][j] = up + left
            return dp[i][j]
        
        return dfs(m-1,n-1)

        # bottom up, tabulation
        dp = [[0 for _ in range(n)] for _ in range(m)]
        dp[0][0] = 1

        for i in range(m):
            for j in range(n):
                if i == 0 and j == 0:
                    continue
                up,left = 0,0
                if i-1 >= 0:
                    up += dp[i-1][j]
                if j-1 >= 0:
                    left += dp[i][j-1]
                dp[i][j] = up + left
        
        print(dp)
        return dp[m-1][n-1]
```

*With obstacles, just a slight change*

```python
class Solution:
    def uniquePathsWithObstacles(self, obstacleGrid: List[List[int]]) -> int:
        m,n = len(obstacleGrid), len(obstacleGrid[0])
        if obstacleGrid[0][0] == 1 or obstacleGrid[m-1][n-1] == 1:
            return 0
        dp = [[0 for _ in range(n)] for _ in range(m)]
        dp[0][0] = 1
        for i in range(m):
            for j in range(n):
                if i == 0 and j == 0:
                    continue
                elif obstacleGrid[i][j] == 1: # obstacle case
                    continue
                
                up,left = 0,0

                if i-1 >= 0:
                    up += dp[i-1][j]
                if j-1 >= 0:
                    left += dp[i][j-1]
                
                dp[i][j] = up+left
        return dp[m-1][n-1]
```

```python
class Solution:
    def minFallingPathSum(self, matrix: List[List[int]]) -> int:
        n = len(matrix)

        dp = [[-1 for _ in range(n)] for _ in range(n)]
        def dfs(i,j):
            
            
            if i<0 or i>=n or j<0 or j>=n:
                return float("inf")
            elif i == n-1:
                return matrix[i][j]
            elif dp[i][j] != -1:
                return dp[i][j]
            
            down = dfs(i+1,j)
            leftd = dfs(i+1,j-1)
            rightd = dfs(i+1,j+1)

            dp[i][j] = min(down,leftd,rightd) + matrix[i][j]
            return dp[i][j]

        # curr_min = float("inf")
        # for j in range(n):
        #     curr_min = min(curr_min, dfs(0,j))
        
        # return curr_min


        for i in range(n-2, -1,-1):
            for j in range(n-1,-1,-1):
                down = matrix[i+1][j]
                downl = float("inf") if j-1 < 0 else matrix[i+1][j-1]
                downr = float("inf") if j+1 >= n else matrix[i+1][j+1]
                matrix[i][j] += min(down,downl,downr) 
        
        curr_min = float("inf")
        for i in range(n):
            curr_min = min(curr_min, matrix[0][i])
        return curr_min
```

**Triangle Path Sum**

```python
class Solution:
    def minimumTotal(self, triangle: List[List[int]]) -> int:
        n = len(triangle)
        dp = [[-1 for _ in range(len(triangle[-1]))] for _ in range(len(triangle))]
        def dfs2(i,j):
            if i == len(triangle)-1:
                return triangle[i][j]
            
            elif dp[i][j] != -1:
                return dp[i][j]
            
            d1 = dfs2(i+1,j)
            d2 = dfs2(i+1,j+1)
            dp[i][j] = min(d1,d2) + triangle[i][j]
            return dp[i][j]
        
        return dfs2(0,0)
```

**New pattern : Partition equals subset Sum**   

```python
class Solution:
    def canPartition(self, nums: List[int]) -> bool:

        totalSum = sum(nums)

        if totalSum % 2 != 0:
            return False

        target = totalSum // 2
        dp = {}

        def backtrack(index, total):
            if total == target:
                return True
            if total > target or index < 0:
                return False
            if (index, total) in dp:
                return dp[(index, total)]

            dp[(index, total)] = (
                backtrack(index - 1, total) or 
                backtrack(index - 1, total + nums[index])
            )
            return dp[(index, total)]

        return backtrack(len(nums) - 1, 0)




class Solution:
    def canPartition(self, nums: List[int]) -> bool:
        total = sum(nums)
        
        # If sum is odd, we can't split it into two equal subsets
        if total % 2 == 1:
            return False
        
        target = total // 2
        n = len(nums)
        

        dp = [[False for _ in range(target+1)] for _ in range(n+1)]


        for i in range(n+1):
            dp[i][0] = True
        

        for i in range(1,n+1):
            for j in range(1,target+1):
                nottake = dp[i-1][j]
                take = False
                if j >= nums[i-1]:
                    take = dp[i-1][j-nums[i-1]]
                
                dp[i][j] = nottake or take
        
        return dp[n][target]
```

Coin Change, min coins required

```python
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:

        dp = [[-1 for _ in range(amount+1)] for _ in range(len(coins))]

        def dfs(i,target):
            if target == 0:
                return 0
            if i == len(coins)-1:
                remainder = target % coins[i]
                if remainder == 0:
                    return target // coins[i]
                return float("inf")
            
            if dp[i][target] != -1:
                return dp[i][target]
            take = float("inf")
            
            if coins[i] <= target:
                take = 1 + dfs(i, target - coins[i])
            nottake = dfs(i+1,target)
            dp[i][target] = min(take, nottake)
            return dp[i][target]
        ans = dfs(0,amount)
        return ans if ans != float("inf") else -1


class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:

        
        dp = {}

        def backtrack(index,target):
            if target == 0:
                return 0
            if target < 0 or index == -1:
                return float("inf")
            
            if (index,target) in dp:
                return dp[(index,target)]

            mini = min(backtrack(index-1,target), 1+backtrack(index,target-coins[index]))
            dp[(index,target)] = mini
            return mini
        
        mini = backtrack(len(coins)-1, amount)
        return  mini if mini != float("inf") else -1 


        
```

```python
class Solution:
    def change(self, amount: int, coins: List[int]) -> int:

        dp = {}

        def backtrack(index,target):
            if target == 0:
                return 1
            elif index == -1 or target < 0:
                return 0
            if (index,target) in dp:
                return dp[(index,target)]
            nottake = backtrack(index-1,target)
            take = backtrack(index,target-coins[index])

            dp[(index,target)] = nottake + take
            return dp[(index,target)]
        
        return backtrack(len(coins)-1, amount)
```

**Longest Common Subsequence**

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:

        dp = [[-1 for _ in range(len(text2)+1)] for _ in range(len(text1)+1)]
        
        
        def dfs(i1,i2):
            if i1 < 0 or i2 < 0:
                return 0
            if dp[i1][i2] != -1:
                return dp[i1][i2]
            
            if text1[i1] == text2[i2]:
                dp[i1][i2] = 1 + dfs(i1-1,i2-1)
                return dp[i1][i2]
            else:
                dp[i1][i2] = max(dfs(i1-1,i2), dfs(i1,i2-1))
                return dp[i1][i2]
        
        # return dfs(len(text1)-1,len(text2)-1)
    
        dp = [[0 for _ in range(len(text2)+1)] for _ in range(len(text1)+1)]
        
        for i in range(1,len(text1)+1):
            for j in range(1,len(text2)+1):
                if text1[i-1] == text2[j-1]:
                    dp[i][j] = 1 + dp[i-1][j-1]
                else:
                    dp[i][j] = max(dp[i-1][j], dp[i][j-1])
        
        return dp[len(text1)][len(text2)]
```

Minimum insertions to make string palindromic

```python
class Solution:
    def minInsertions(self, s: str) -> int:

        t = s[::-1]
        if s == t:
            return 0
        dp = [[0 for _ in range(len(t)+1)] for _ in range(len(s)+1)]



        for i in range(1,len(s)+1):
            for j in range(1,len(t)+1):
                if s[i-1] == t[j-1]:
                    dp[i][j] = 1 + dp[i-1][j-1]
                else:
                    dp[i][j] = max(dp[i-1][j], dp[i][j-1])
        
        return len(s) - dp[len(s)][len(t)]
```

```python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:

        dp = [[-1 for _ in range(len(word2))] for _ in range(len(word1))]

        def dfs(i1,i2):
            if i1 < 0:
                return i2 + 1
            if i2 < 0:
                return i1 + 1
            
            if dp[i1][i2] != -1:
                return dp[i1][i2]


            if word1[i1] == word2[i2]:
                dp[i1][i2] = dfs(i1-1,i2-1)
                return dp[i1][i2]
            
            dp[i1][i2] = 1 + min(dfs(i1-1,i2),dfs(i1,i2-1),dfs(i1-1,i2-1))
            return dp[i1][i2]



        # return dfs(len(word1)-1, len(word2)-1)
        dp = [[0 for _ in range(len(word2)+1)] for _ in range(len(word1)+1)]

        for i in range(len(word1)+1):
            dp[i][0] = i

        for i in range(len(word2)+1):
            dp[0][i] = i

        for i in range(1,len(word1) + 1):
            for j in range(1,len(word2) + 1):
                if word1[i-1] == word2[j-1]:
                    dp[i][j] = dp[i-1][j-1]
                else:
                    dp[i][j] = 1 + min(dp[i-1][j], dp[i-1][j-1],dp[i][j-1])

        return dp[-1][-1]
```

```python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        
        n1,n2 = len(word1), len(word2)
        dp = {}

        def backtrack(index1,index2):
            if index1 < 0:
                return index2+1
            elif index2 < 0:
                return index1+1
            
            if (index1,index2) in dp:
                return dp[(index1,index2)]

            if word1[index1] == word2[index2]:
                dp[(index1,index2)] = backtrack(index1-1,index2-1)
                return dp[(index1,index2)]
            
            addOrRemove = 1 + min(backtrack(index1-1,index2),backtrack(index1,index2-1))
            replace = 1 + backtrack(index1-1,index2-1) 
            
            dp[(index1,index2)] = min(addOrRemove, replace)
            return dp[(index1,index2)]
        
        return backtrack(n1-1, n2-1)
```

LongestIncreasingSubsequence

```python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:

        dp = [-1] * len(nums)

        def backtrack(index):
            if dp[index] != -1:
                return dp[index]

            maxi = 1
            for i in range(index+1,len(nums)):
                if nums[i] > nums[index]:
                    maxi = max(maxi,1+backtrack(i))
            dp[index] = maxi
            return dp[index]
        
        maxi = 0
        for i in range(len(nums)):
            maxi = max(backtrack(i), maxi)
        return maxi
        
```

```python
# start from popping the last balloon, go to the top 

class Solution:
    def maxCoins(self, nums: List[int]) -> int:
        nums = [1] + nums + [1]

        dp = [[-1 for _ in range(len(nums))] for _ in range(len(nums))]

        def dfs(i,j):
            if i>j:
                return 0
            
            if dp[i][j] != -1:
                return dp[i][j]
            
            max_coins = float("-inf")
            for index in range(i,j+1):
                tmp = (nums[i-1] * nums[index] * nums[j+1]) + dfs(i,index-1) + dfs(index+1,j)
                max_coins = max(max_coins,tmp)
            
            dp[i][j] = max_coins
            return dp[i][j]
        
        return dfs(1,len(nums)-2)

        
```
