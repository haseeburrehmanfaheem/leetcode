```python
# single number
def single_number(nums):
    
# Xor 
# 1 ^ 0 : 1
# 0 ^ 1 : 1
# 1 ^ 1 : 0
# 0 ^ 0 : 0
# n ^ 0 : n
# n ^ n : 0

    ans = 0
    for n in nums:
        ans ^= n
    return ans 
```

```python
class Solution:
    def countBits(self, n: int) -> List[int]:

        dp = [0] * (n+1)

        for i in range(1,n+1):
            rightshift = i >> 1
            # if number is odd
            # it has a leading 1
            # we can rightshift, get the answer from dp and add 1
            if i & 1 != 0: 
                #odd
                dp[i] = dp[rightshift] + 1
            else:
                dp[i] = dp[rightshift]
        
        return dp

        
```

```python
class Solution:
    def reverseBits(self, n: int) -> int:
        reversed = 0

        for _ in range(32): # since the number is a 32 bit number
            lastbit = n & 1
            n = n >> 1
            reversed = reversed << 1
            reversed = lastbit | reversed
        
        return reversed
        
```

```python
class Solution:
    def reverse(self, x: int) -> int:
        ans = 0
        is_neg = False
        if x < 0:
            is_neg = True
            x = x * -1
        while x:
            last = x%10
            ans = ans * 10
            ans = ans + last
            x = x // 10
        ans = ans if not is_neg else ans * -1
        INT_MIN = -2**31
        INT_MAX = 2**31 - 1

        if ans < INT_MIN or ans > INT_MAX:
            return 0
        return ans
        
```
