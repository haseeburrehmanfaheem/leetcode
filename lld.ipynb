Encapsulation : Hiding the implementation details of an object from the outside world and only allowing access through public setters and getters
prevents misuse

class BankAccount {
    // Private attributes
    private String accountNumber;
    private double balance;

    // Constructor to initialize account number and balance
    public BankAccount(String accountNumber, double balance) {
        this.accountNumber = accountNumber;
        this.balance = balance;
    }

    // Method to deposit an amount
    public void deposit(double amount) {
        balance += amount;
    }

    // Method to withdraw an amount if sufficient balance is available
    public void withdraw(double amount) {
        if (balance >= amount) {
            balance -= amount;
        } else {
            System.out.println("Insufficient funds.");
        }
    }

    // Method to get the balance
    public double getBalance() {
        return balance;
    }
}

Inheritance allows a class to inherit properties and methods from another class, called a superclass or parent class

Class Vehicle{
    String color;

    public Vehicle(String color){
        this.color = color;
    }

    public displayColor(){
        System.out.println("The vehicle has a " + this.color + " color");
    }
}

Class Car extends Vehicle{
    int speed;

    public Car(int speed, String color){
        super(color);
        this.speed = speed;
    }
    public void accelerate() {
        speed += 10;
    }
    public void displaySpeed() {
        System.out.println("Current speed: " + speed + " km/h");
    }
}

Polymorphism is the ability of an object to take on multiple forms.
Compile-Time Polymorphism (Method Overloading)
methods have same name but different parameters

class MathOperations {
    // Overloaded method with two int parameters
    public int add(int a, int b) {
        return a + b;
    }

    // Overloaded method with three int parameters
    public int add(int a, int b, int c) {
        return a + b + c;
    }
}

Run-Time Dynamic Polymorphism (Method Overriding)
Happens when a subclass provides specific implementation of a method that is already defined in superclass

// Parent class
class Animal {
    // Method to be overridden
    public void makeSound() {
        System.out.println("Animal makes a sound");
    }
}

// Child class
class Dog extends Animal {
    // Overriding the parent class method
    @Override
    public void makeSound() {
        System.out.println("bark bark");
    }
}


Abstraction is the concept of showing only the necessary information to the outside world while hiding unnecessary details.

abstract class Animal {
    // Abstract method (without implementation)
    abstract void makeSound();

    // Concrete method (with implementation)
    public void sleep() {
        System.out.println("Sleeping...");
    }
}

// Subclass 1
class Dog extends Animal {
    @Override
    void makeSound() {
        System.out.println("Dog barks");
    }
}

// Subclass 2
class Cat extends Animal {
    @Override
    void makeSound() {
        System.out.println("Cat meows");
    }
}

S.O.L.I.D
A class should have single responsibility
Open/closed : class should be open to extension, but closed to modification
LSQ : superclass' object should be replacable with objects of its subclass
ISP : No client should be forced to depend on interfaces they don't use., break interfaces down into smaller ones
High-level modules should not depend on low-level modules; both should depend on abstractions.
